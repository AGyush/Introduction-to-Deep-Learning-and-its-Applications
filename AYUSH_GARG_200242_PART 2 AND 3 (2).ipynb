```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv("House_prediction.csv")
plt.scatter(data['area'], data['rent amount (R$)'])
plt.show()
```

In this we can see that the area of houses is condensed in a small region exceot for 3-4 cases in which the area is abnormally high but the rent for them is still not very high. 
On the other hand we can see that in the limited range of area there is a huge variation in rent amount. Some houses have abnormally high rent amounts for the same area range while most of them are again limited to a range.

```python
plt.scatter(data['rooms'], data['fire insurance (R$)'])
plt.show()
```

In this the graph between no. of rooms and the fire insurance, we can see that the average value of fire insurance remains almost the same with increase in the no. of rooms. It becomes a bit less uniform as no. of rooms increase but that might be beacuse of the small subset of larger rooms. There are also a few exceptions where the fire insurance is exceptionally high.

```python
plt.scatter(data['furniture'], data['fire insurance (R$)'])
plt.show()
```

Again except a few abnormalities in the graph,the fire insurance does not show much dependance upon if the flat is furnished or not.

```python
plt.scatter(data['parking spaces'], data['fire insurance (R$)'])
plt.show()
```

Again the average value of fire insurance does not show much variation with change in no. of parking spaces.


```python
plt.scatter(data['floor'], data['fire insurance (R$)'])
plt.show()
```

```python
plt.scatter(data['rent amount (R$)'], data['fire insurance (R$)'])
plt.show()
```

Finally we get a sort of a linear graph. This shows that fire insurance is more or less dependent on the rent amount only.


```python
plt.scatter(data['furniture'], data['property tax (R$)'])
plt.show()
```

This shows that non furnished apartments usually have high propert tax rates


```python
plt.scatter(data['rooms'], data['property tax (R$)'])
plt.show()
```

However no. of rooms do not seem to have a linear relationship with the property tax


```python
plt.scatter(data['floor'], data['property tax (R$)'])
plt.show()
```

Similarly no. of floors do not seem to affect the property tax much.
